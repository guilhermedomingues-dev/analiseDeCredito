```python
from google.colab import output

print("Login")
nome=str(input("Seu nome: ")).capitalize()
output.clear()

print("Olá,",nome)
valor=float(input("Qual valor deseja de crédito? R$"))
score=int(input("Seu Score: "))
output.clear()

if score<=300:
  print(nome,", seu crédito no valor de R$", valor, " foi negado!")
elif 300<score<500:
  tempo=int(input("Qual o tempo, em meses, de inatividade da sua conta corrente: "))
  output.clear()

  if tempo == 0:
    print(nome,", seu crédito no valor de R$", valor, " foi aprovado!")
  elif 3<tempo<5:
    renda=float(input("Sua renda mensal: R$"))
    output.clear()

    validacao=0.3*renda
    if valor>validacao:
      print(nome,", seu crédito no valor de R$", valor, " foi negado!")
    elif valor<validacao:
      print(nome,", seu crédito no valor de R$", valor, " foi aprovado!")
    else:
      print("Erro!")

  elif tempo<=6:
    print(nome,", sua conta está inativa.\nSeu crédito no valor de R$", valor, " foi negado!")
  else:
    print("Erro!")

elif 500<score<700:
  trabalho=int(input("Em meses, seu tempo de trabalho: "))
  output.clear()

  if trabalho < 6:
    print(nome,", seu crédito no valor de R$", valor, " foi negado!")
  elif trabalho >= 6:
    print(nome,", seu crédito no valor de R$", valor, " foi aprovado!")
  else:
    print("Erro!")

elif 700<score<1000:
  print(nome,", seu crédito no valor de R$", valor, " foi aprovado!")
else:
  print("Erro!")
```
